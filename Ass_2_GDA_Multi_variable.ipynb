```python
import sympy as sym
from sympy import *
```

```python
import random
```

```python
#step-1 intialize
X = random.randint(1,5)
Y = random.randint(1,5)
print("Random values of X = {}".format(X)+" Y = {}".format(Y))
n = 0.01
epochs = 150
iter = 1
x=0
x = sym.Symbol('x')
y = sym.Symbol('y')
print("Customize the funtion f = a*x**i+b*y**j+c")
a = int(input('Enter coefficient of x (if exists or else 0) a:'))
i = int(input('Enter the power of x (if exists or else 0) i:'))
b = int(input('Enter coefficient of y (if exists or else 0) b:'))
j = int(input('Enter the power of y (if exists or else 0) j:'))
c = int(input('Enter constants if exist c:'))
f = a*x**i+b*y**j+c
print("INPUT FUNCTION f(x,y): {}".format(f))
print("---------------------------------------------------")
print("Step 1: Initialization done and function is {}".format(f))
while 1:
  print("Iteration number: {}".format(iter))
  #step 2
  #Calculate Gradient/slope/firstorderderivative
  #derivative wrto x
  dfx = sym.diff(f,x)
  dfy = sym.diff(f,y) #derivative wrto y
  slopex = dfx.subs(x,X) #after substitution of X in slope (dfx)
  slopey = dfy.subs(y,Y) #after substitution of Y in slope (dfy)
  print("Step 2: slopex = {}".format(slopex)+" slopey = {}".format(slopey))
  #step 3 - find change in variables values - we get step length
  delta_X = -(n)*slopex
  delta_Y = -(n)*slopey
  print("Step 3: delta_X {}".format(delta_X)+" delta_Y {}".format(delta_Y))
  #step 4: updating variable value
  X = X+delta_X
  Y = Y+delta_Y
  print("Step 4: Updated Value X = {}".format(X)+" Y = {}".format(Y))

  resf = f.subs(x,X)
  res = resf.subs(y,Y)
  print("Checking for the result value of the function f(updated_X,updated_Y) = {}".format(res))

  #step 5: update iter
  print("step 5: Update iteration")
  iter = iter+1
  if(iter>epochs):
    break;
  else:
    print("Goto step 2")
  print("-------------------------------------------------------------------")

print("---------------------------------------------------------------------------")
print("Global Minima point obtained: after rounding to two values after decimal point")
print("X: ",round(X,2))
print("Y: ",round(Y,2))
optimalres = f.subs(x,X)
optimalfinal = optimalres.subs(y,Y)
print("Optimal function value: ",round(optimalfinal,2))
```
