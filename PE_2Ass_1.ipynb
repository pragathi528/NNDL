```
import random
from sympy import *
x=symbols('x')
f=x**4+3*x**2+10
y=diff(f,x)
print(y)
for i in range(1,5):
  a=random.randint(1,5)
print(a)
itr=1
epox=100
n=0.1
while(itr<=epox):
  Y=y.subs(x,a)
  print("y=",Y)
  f1=-n*Y
  print("f1(x)=",f1)
  x1=a+f1
  print("x=",x1)
  itr=itr+1
  a=x1
  print(" ")
print("Final x=",a)
print("Value is ",f.subs(x,a))
```
